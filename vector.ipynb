```python
class Node(object):
    
    def __init__(self, name, value, children=[]):
        self.name = name
        self.value = value
        self.children = children
        self.leaves = len(children)

    def __repr__(self):
        return self.name + " " + np.array2string(self.value)

def javalang_to_nodes_recursive(javalang_node, node_list,n_feature=10):
    #print(type(javalang_node))
    if issubclass(type(javalang_node), javalang.ast.Node):
        new_node = Node(name=str(type(javalang_node)),value=np.random.uniform(low=-1,high=1,size=(n_feature,)), children=[])
        node_list.append(new_node)
        for child in javalang_node.children:
            new_node.children.append(javalang_to_nodes_recursive(child, node_list, n_feature))
    else:
        new_node = Node(name=str(javalang_node),value=np.random.uniform(low=-1,high=1,size=(n_feature,)), children=[])
        node_list.append(new_node)
    return new_node

def javalang_to_nodes_iterative(javalang_node, node_dict, n_feature=10):

    for path, node in javalang_node:
        new_node = Node(name=str(type(node)),value=np.random.uniform(low=-1,high=1,size=(n_feature,)), children=[])
        node_dict.update({node.position:new_node})

        if len(path) > 0:
            if issubclass(type(path[-1]), javalang.ast.Node):
                node_dict[path[-1].position].children.append(new_node)
            elif issubclass(type(path[-2]), javalang.ast.Node):
                node_dict[path[-2].position].children.append(new_node)
            #else:
                #print("no parent")
        
    return
```

```python
def ssr_gradient(x, y, b):
    res = b[0] + b[1] * x - y
    return res.mean(), (res * x).mean()

# find the vectorized representation according to this node's children
def vec(node, wl, wr, b):
    children_sum = 0
    if node.leaves == 0:
        return node.value
    for child in node.children:
        i = node.children.index(child)
        n = node.leaves
        wi = (n-i)/(n-1)*wl + (i-1)/(n-1)*wr
        children_sum += child.leaves/n * wi.dot(child.value)
    return np.tanh(children_sum + b)

# find the distance between this node's VR now and its VR according to its children
# negative sampling added 
def distance(tree, node, wl, wr, b):
    margin = 1

    d = np.linalg.norm(node.value-vec(node,wl,wr,b))**2

    random_symbol_key, random_symbol = random.choice(list(tree.items()))
    random_sample_key, random_sample = random.choice(list(tree.items()))
    dx = np.linalg.norm(random_symbol.value-vec(random_sample,wl,wr,b))**2

    return max(0, margin + d - dx)

# find the gradient of weight left, weight right and bias
def cvr_gradient(tree, batch, wl,wr, b):
    dwl, dwr, db = 0, 0, 0

    # will update lambda if overfitting observed
    balance_lambda = 0
    n_features = 10 

    # for every node in the tree we sum up the gradient and take average at the end
    for key in batch:

        # if the distance + margin is larger than negative sample
        # then gradient of weights and bias is not 0
        if distance(tree, tree[key], wl, wr, b) > 0:
            
            child_dwl, child_dwr, child_db = 0, 0, 0
            # go through children of this node 
            for child in tree[key].children:
                if child.leaves == 0:
                    break
                
                i = tree[key].children.index(child)
                n = tree[key].leaves
                vec_from_children = (n-i)/(n-1) * wl.dot(child.value) + (i-1)/(n-1) * wr.dot(child.value) + b

                # should the negative sample be part of gradient?

                # d/dwl
                child_dwl += -tree[key].value * (n-i)/(n-1) * (tree[key].value - np.tanh(vec_from_children))*(1 - np.tanh(vec_from_children)**2)

                # d/dwr
                child_dwr += -tree[key].value * (i-1)/(n-1) * (tree[key].value - np.tanh(vec_from_children))*(1 - np.tanh(vec_from_children)**2)

                # d/db
                child_db += (tree[key].value - np.tanh(vec_from_children)) * (np.tanh(vec_from_children)**2 - 1)

            if tree[key].leaves != 0:
                dwl += child_dwl/node.leaves
                dwr += child_dwr/node.leaves
                db += child_db/node.leaves

    # l2 regularization
    dwl += balance_lambda*np.linalg.norm(wl)/(2*n_features**2)
    dwr += balance_lambda*np.linalg.norm(wr)/(2*n_features**2)
    

    return dwl/len(batch), dwr/len(batch), db/len(batch)

def gradient_descent(
    gradient, x, y, start, learn_rate=0.1, n_iter=50, tolerance=1e-06
):
    vector = start
    for _ in range(n_iter):
        diff = -learn_rate * np.array(gradient(x, y, vector))
        if np.all(np.abs(diff) <= tolerance):
            break
        vector += diff
    return vector
```

```python
def sgd(
    gradient, tree, wl, wr, b, start, learn_rate=0.1, decay_rate=0.0, batch_size=1,
    n_iter=50, tolerance=1e-06, dtype="float64", random_state=None
):
    # Checking if the gradient is callable
    if not callable(gradient):
        raise TypeError("'gradient' must be callable")

    # Setting up the data type for NumPy arrays
    dtype_ = np.dtype(dtype)

    # Converting x and y to NumPy arrays
    #x, y = np.array(x, dtype=dtype_), np.array(y, dtype=dtype_)
    #n_obs = x.shape[0]
    n_obs = len(tree)
    #if n_obs != y.shape[0]:
    #    raise ValueError("'x' and 'y' lengths do not match")
    #xy = np.c_[x.reshape(n_obs, -1), y.reshape(n_obs, 1)]

    # Initializing the random number generator
    seed = None if random_state is None else int(random_state)
    rng = np.random.default_rng(seed=seed)

    # Initializing the values of the variables
    # vector = np.array(start, dtype=dtype_)

    # Setting up and checking the learning rate
    learn_rate = np.array(learn_rate, dtype=dtype_)
    if np.any(learn_rate <= 0):
        raise ValueError("'learn_rate' must be greater than zero")

    # Setting up and checking the decay rate
    decay_rate = np.array(decay_rate, dtype=dtype_)
    if np.any(decay_rate < 0) or np.any(decay_rate > 1):
        raise ValueError("'decay_rate' must be between zero and one")

    # Setting up and checking the size of minibatches
    batch_size = int(batch_size)
    if not 0 < batch_size <= n_obs:
        raise ValueError(
            "'batch_size' must be greater than zero and less than "
            "or equal to the number of observations"
        )

    # Setting up and checking the maximal number of iterations
    n_iter = int(n_iter)
    if n_iter <= 0:
        raise ValueError("'n_iter' must be greater than zero")

    # Setting up and checking the tolerance
    tolerance = np.array(tolerance, dtype=dtype_)
    if np.any(tolerance <= 0):
        raise ValueError("'tolerance' must be greater than zero")

    # Setting the difference to zero for the first iteration
    diff = 0

    # Performing the gradient descent loop
    for _ in range(n_iter):
        # Shuffle x and y
        #rng.shuffle(xy)

        # Performing minibatch moves
        for start in range(0, n_obs, batch_size):
            stop = start + batch_size
            # x_batch, y_batch = xy[start:stop, :-1], xy[start:stop, -1:]
            tree_batch = {key: value for key, value in tree.items() if list(tree.keys()).index(key) >= start and list(tree.keys()).index(key) < stop}
            #tree_batch = tree[start:stop]

            # Recalculating the difference
            # grad = np.array(gradient(x_batch, y_batch, vector), dtype_)
            grad = np.array(gradient(tree, tree_batch, wl, wr, b), dtype_)
            #print(grad)
            diff = decay_rate * diff - learn_rate * grad

            # Checking if the absolute difference is small enough
            if np.all(np.abs(diff) <= tolerance):
                break

            # Updating the values of the variables
            wl += diff[0]
            wr += diff[1]
            b += diff[2]

            for key in tree:
                tree[key].value = vec(tree[key], wl, wr, b)

    # return vector if vector.shape else vector.item()
    return tree, wl, wr, b
```
